```python
"""
AI-Driven Financial Management Chatbot - Complete Prototype
Uses IBM Granite 3.2-2B Instruct Model
Designed to run on Google Colab

INSTRUCTIONS TO RUN:
1. Open Google Colab (colab.research.google.com)
2. Create a new notebook
3. Copy this entire code
4. Run the installation cell first (uncomment the pip install line)
5. Run the main code cells
6. Use the interactive interface at the bottom

Author: Financial AI Assistant
Date: November 2025
"""

# ============================================================================
# INSTALLATION CELL - RUN THIS FIRST IN COLAB
# ============================================================================

# Uncomment and run this in the first cell of your Colab notebook:
"""
!pip install -q transformers torch accelerate huggingface_hub pandas matplotlib seaborn plotly
"""

# ============================================================================
# IMPORTS
# ============================================================================

import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
import json
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Dict, List, Optional, Tuple
import warnings
warnings.filterwarnings('ignore')

# ============================================================================
# CORE AI MODEL CLASS
# ============================================================================

class GraniteFinancialAdvisor:
    """Main AI Financial Advisor using IBM Granite 3.2-2B"""

    def __init__(self):
        """Initialize the IBM Granite model and data structures"""
        print("=" * 60)
        print("üöÄ Initializing AI Financial Advisor")
        print("=" * 60)
        print("üì¶ Loading IBM Granite 3.2-2B Instruct Model...")
        print("‚è≥ First-time setup may take 2-3 minutes...")

        model_name = "ibm-granite/granite-3.2-2b-instruct"

        try:
            # Load tokenizer
            self.tokenizer = AutoTokenizer.from_pretrained(model_name)

            # Load model with optimizations
            self.model = AutoModelForCausalLM.from_pretrained(
                model_name,
                device_map="auto",
                torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
                low_cpu_mem_usage=True
            )

            print("‚úÖ Model loaded successfully!")
            print(f"üíª Running on: {'GPU' if torch.cuda.is_available() else 'CPU'}")

        except Exception as e:
            print(f"‚ùå Error loading model: {e}")
            print("üí° Tip: Make sure you have internet connection")
            raise

        # Initialize data storage
        self.user_profile = {}
        self.family_members = []
        self.expenses_log = []
        self.income_log = []
        self.goals = []
        self.bills_due = []
        self.conversation_history = []
        self.spending_patterns = {}
        self.gamification_points = 0
        self.challenges = []

        print("=" * 60)

    def generate_response(self, prompt: str, context: str = "", max_tokens: int = 300) -> str:
        """Generate AI response using Granite model"""

        # Build the full prompt with context
        if context:
            full_prompt = f"{context}\n\nUser Question: {prompt}\n\nFinancial Advisor:"
        else:
            full_prompt = f"You are a helpful financial advisor. Answer this question:\n\n{prompt}\n\nAnswer:"

        # Tokenize
        inputs = self.tokenizer(full_prompt, return_tensors="pt", truncation=True, max_length=1024)

        # Move to same device as model
        inputs = {k: v.to(self.model.device) for k, v in inputs.items()}

        # Generate
        with torch.no_grad():
            outputs = self.model.generate(
                **inputs,
                max_new_tokens=max_tokens,
                temperature=0.7,
                top_p=0.9,
                do_sample=True,
                pad_token_id=self.tokenizer.eos_token_id
            )

        # Decode response
        response = self.tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Extract only the generated part (after the prompt)
        response = response.split("Answer:")[-1].strip()
        if "User Question:" in response:
            response = response.split("User Question:")[0].strip()

        return response

    def analyze_with_ai(self, query: str, data: Dict) -> str:
        """Use AI to analyze user data and provide insights"""

        context = f"""Financial Profile Summary:
- Category: {data.get('category', 'Unknown')}
- Monthly Budget: ‚Çπ{data.get('budget', 0):,.0f}
- Total Income: ‚Çπ{data.get('total_income', 0):,.0f}
- Total Expenses: ‚Çπ{data.get('total_expenses', 0):,.0f}
- Savings: ‚Çπ{data.get('savings', 0):,.0f}
- Location: {data.get('location', 'Unknown')}
- Goals: {', '.join(data.get('goals', ['None']))}

Based on this financial data, provide specific, actionable advice."""

        return self.generate_response(query, context)

# ============================================================================
# USER ONBOARDING & PROFILE MANAGEMENT
# ============================================================================

class UserProfileManager:
    """Manage user profiles and onboarding"""

    DEMOGRAPHICS = {
        'student': ['undergraduate', 'postgraduate', 'school_student'],
        'employed': ['private_sector', 'government', 'self_employed'],
        'unemployed': ['job_seeking', 'career_break'],
        'housewife': ['homemaker'],
        'retired': ['senior_citizen']
    }

    MARITAL_STATUS = ['single', 'married', 'divorced', 'widowed']

    LIVING_SITUATIONS = ['own_home', 'rented', 'pg', 'hostel', 'with_parents']

    INDIAN_CITIES_COST_INDEX = {
        'mumbai': 1.0, 'delhi': 0.95, 'bangalore': 0.90, 'hyderabad': 0.75,
        'chennai': 0.75, 'kolkata': 0.70, 'pune': 0.85, 'ahmedabad': 0.70,
        'jaipur': 0.65, 'lucknow': 0.60, 'mysuru': 0.65, 'tier2': 0.60, 'tier3': 0.50
    }

    @staticmethod
    def conduct_onboarding() -> Dict:
        """Interactive onboarding questionnaire"""
        print("\n" + "=" * 60)
        print("üëã WELCOME TO AI FINANCIAL ADVISOR")
        print("=" * 60)
        print("Let's set up your profile (this is one-time setup)\n")

        profile = {}

        # Basic Information
        profile['name'] = input("üìù Your name: ").strip()
        profile['age'] = int(input("üéÇ Your age: "))

        # Demographics
        print("\nüìä Select your category:")
        categories = ['student', 'employed', 'unemployed', 'housewife', 'retired']
        for i, cat in enumerate(categories, 1):
            print(f"  {i}. {cat.replace('_', ' ').title()}")
        cat_choice = int(input("Enter choice (1-5): ")) - 1
        profile['category'] = categories[cat_choice]

        # Marital Status
        print("\nüíë Marital status:")
        for i, status in enumerate(UserProfileManager.MARITAL_STATUS, 1):
            print(f"  {i}. {status.title()}")
        marital_choice = int(input("Enter choice: ")) - 1
        profile['marital_status'] = UserProfileManager.MARITAL_STATUS[marital_choice]

        # Family Size
        profile['family_size'] = int(input("\nüë®‚Äçüë©‚Äçüëß‚Äçüë¶ Total family members (including you): "))
        profile['dependents'] = int(input("üë∂ Number of dependents (children, elderly): "))

        # Location
        print("\nüìç Location (for cost of living analysis):")
        profile['city'] = input("Enter your city: ").strip().lower()
        profile['country'] = input("Enter your country (default: India): ").strip() or "India"

        # Living Situation
        print("\nüè† Living situation:")
        for i, situation in enumerate(UserProfileManager.LIVING_SITUATIONS, 1):
            print(f"  {i}. {situation.replace('_', ' ').title()}")
        living_choice = int(input("Enter choice: ")) - 1
        profile['living_situation'] = UserProfileManager.LIVING_SITUATIONS[living_choice]

        # Assets
        print("\nüöó Assets:")
        profile['owns_vehicle'] = input("Do you own a vehicle? (yes/no): ").lower() == 'yes'
        if profile['owns_vehicle']:
            profile['vehicle_type'] = input("Vehicle type (car/bike/both): ").lower()

        profile['owns_property'] = input("Do you own property? (yes/no): ").lower() == 'yes'

        # Financial Goals
        print("\nüéØ Financial Goals & Dreams:")
        print("(Enter your goals one by one, press Enter without text when done)")
        goals = []
        while True:
            goal = input(f"  Goal {len(goals)+1}: ").strip()
            if not goal:
                break
            goals.append(goal)
        profile['goals'] = goals

        profile['created_at'] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        print("\n‚úÖ Profile created successfully!")
        return profile

    @staticmethod
    def get_cost_of_living_multiplier(city: str) -> float:
        """Get cost of living multiplier for a city"""
        city = city.lower()
        return UserProfileManager.INDIAN_CITIES_COST_INDEX.get(city, 0.60)

# ============================================================================
# INCOME & EXPENSE TRACKING
# ============================================================================

class FinancialTracker:
    """Track income and expenses"""

    INCOME_CATEGORIES = {
        'student': ['pocket_money', 'scholarship', 'part_time_job', 'tuition', 'freelancing'],
        'employed': ['salary', 'bonus', 'investment_returns', 'rental_income', 'side_business'],
        'unemployed': ['savings', 'family_support', 'freelancing', 'unemployment_benefits'],
        'housewife': ['family_income', 'side_business', 'investment_returns'],
        'retired': ['pension', 'investment_returns', 'rental_income', 'savings']
    }

    EXPENSE_CATEGORIES = [
        'food_groceries', 'rent', 'utilities', 'transportation', 'healthcare',
        'education', 'entertainment', 'clothing', 'insurance', 'emi_loans',
        'communication', 'personal_care', 'household_items', 'gifts_donations',
        'travel', 'investments', 'emergency_fund', 'other'
    ]

    @staticmethod
    def collect_income(user_category: str) -> List[Dict]:
        """Collect income details based on user category"""
        print("\n" + "=" * 60)
        print("üí∞ INCOME COLLECTION")
        print("=" * 60)

        applicable_sources = FinancialTracker.INCOME_CATEGORIES.get(user_category,
                            FinancialTracker.INCOME_CATEGORIES['employed'])

        print(f"\nApplicable income sources for {user_category}:")
        for i, source in enumerate(applicable_sources, 1):
            print(f"  {i}. {source.replace('_', ' ').title()}")

        income_list = []
        print("\nEnter your income sources (press 0 when done):")

        while True:
            choice = int(input("\nSelect income source (0 to finish): "))
            if choice == 0:
                break

            if 1 <= choice <= len(applicable_sources):
                source = applicable_sources[choice - 1]
                amount = float(input(f"Monthly amount for {source.replace('_', ' ')}: ‚Çπ"))
                frequency = input("Frequency (monthly/yearly/one-time): ").lower()

                income_list.append({
                    'source': source,
                    'amount': amount,
                    'frequency': frequency,
                    'date': datetime.now().strftime("%Y-%m-%d")
                })
                print(f"‚úÖ Added: ‚Çπ{amount:,.0f} from {source}")

        total = sum(inc['amount'] for inc in income_list)
        print(f"\nüíµ Total Monthly Income: ‚Çπ{total:,.0f}")

        return income_list

    @staticmethod
    def collect_expenses_interactive() -> List[Dict]:
        """Interactive expense collection with smart categorization"""
        print("\n" + "=" * 60)
        print("üí≥ EXPENSE COLLECTION")
        print("=" * 60)
        print("\nüìù Smart Expense Entry - Just describe your expense!")
        print("Example: 'Groceries at supermarket 2500'")
        print("         'Electricity bill 1500'")
        print("         'Movie tickets 800'")
        print("\nType 'done' when finished\n")

        expenses = []

        while True:
            entry = input("üí∏ Enter expense: ").strip()

            if entry.lower() == 'done':
                break

            # Try to parse the entry
            parts = entry.split()
            try:
                # Last part should be amount
                amount = float(parts[-1])
                description = ' '.join(parts[:-1])

                # Auto-categorize based on keywords
                category = FinancialTracker._auto_categorize(description)

                print(f"   Category: {category.replace('_', ' ').title()}")
                confirm = input("   Correct? (y/n): ").lower()

                if confirm != 'y':
                    print("\n   Available categories:")
                    for i, cat in enumerate(FinancialTracker.EXPENSE_CATEGORIES, 1):
                        print(f"   {i}. {cat.replace('_', ' ').title()}")
                    cat_choice = int(input("   Select category: ")) - 1
                    category = FinancialTracker.EXPENSE_CATEGORIES[cat_choice]

                expenses.append({
                    'description': description,
                    'amount': amount,
                    'category': category,
                    'date': datetime.now().strftime("%Y-%m-%d"),
                    'time': datetime.now().strftime("%H:%M:%S")
                })

                print(f"   ‚úÖ Added: ‚Çπ{amount:,.0f} - {description}\n")

            except (ValueError, IndexError):
                print("   ‚ö†Ô∏è  Invalid format. Try: 'description amount'")

        total = sum(exp['amount'] for exp in expenses)
        print(f"\nüí∞ Total Expenses: ‚Çπ{total:,.0f}")

        return expenses

    @staticmethod
    def _auto_categorize(description: str) -> str:
        """Auto-categorize expense based on description keywords"""
        description = description.lower()

        keywords = {
            'food_groceries': ['grocery', 'food', 'vegetables', 'supermarket', 'provisions'],
            'rent': ['rent', 'lease', 'house rent'],
            'utilities': ['electricity', 'water', 'gas', 'internet', 'phone bill'],
            'transportation': ['petrol', 'diesel', 'fuel', 'bus', 'taxi', 'uber', 'ola', 'auto'],
            'healthcare': ['medicine', 'doctor', 'hospital', 'health', 'medical'],
            'education': ['school', 'college', 'tuition', 'books', 'course', 'fees'],
            'entertainment': ['movie', 'restaurant', 'dining', 'party', 'gaming', 'netflix'],
            'clothing': ['clothes', 'shoes', 'dress', 'shirt', 'fashion'],
        }

        for category, words in keywords.items():
            if any(word in description for word in words):
                return category

        return 'other'

# ============================================================================
# BUDGET MANAGEMENT & ANALYSIS
# ============================================================================

class BudgetAnalyzer:
    """Analyze budget and provide recommendations"""

    @staticmethod
    def analyze_budget(income: float, expenses: List[Dict], user_profile: Dict) -> Dict:
        """Comprehensive budget analysis"""

        total_expenses = sum(exp['amount'] for exp in expenses)
        savings = income - total_expenses
        savings_rate = (savings / income * 100) if income > 0 else 0

        # Categorize expenses
        category_totals = {}
        for exp in expenses:
            cat = exp['category']
            category_totals[cat] = category_totals.get(cat, 0) + exp['amount']

        # Calculate recommended percentages (50-30-20 rule adapted)
        needs_budget = income * 0.50  # Essential needs
        wants_budget = income * 0.30  # Wants
        savings_budget = income * 0.20  # Savings/Investments

        # Classify categories into needs/wants
        needs_categories = ['rent', 'utilities', 'food_groceries', 'healthcare',
                          'transportation', 'insurance', 'emi_loans']

        actual_needs = sum(category_totals.get(cat, 0) for cat in needs_categories)
        actual_wants = total_expenses - actual_needs

        analysis = {
            'total_income': income,
            'total_expenses': total_expenses,
            'savings': savings,
            'savings_rate': savings_rate,
            'category_breakdown': category_totals,
            'needs_vs_wants': {
                'needs_actual': actual_needs,
                'needs_budget': needs_budget,
                'wants_actual': actual_wants,
                'wants_budget': wants_budget,
                'savings_actual': savings,
                'savings_budget': savings_budget
            },
            'status': 'surplus' if savings > 0 else 'deficit',
            'deficit_amount': abs(savings) if savings < 0 else 0
        }

        return analysis

    @staticmethod
    def suggest_adjustments(analysis: Dict, ai_advisor: GraniteFinancialAdvisor) -> List[str]:
        """AI-powered budget adjustment suggestions"""

        suggestions = []

        # Rule-based suggestions
        if analysis['status'] == 'deficit':
            deficit = analysis['deficit_amount']
            suggestions.append(f"‚ö†Ô∏è Budget Deficit: You're overspending by ‚Çπ{deficit:,.0f}")

            # Find highest expense categories
            sorted_expenses = sorted(
                analysis['category_breakdown'].items(),
                key=lambda x: x[1],
                reverse=True
            )

            suggestions.append("\nüìâ Cost-cutting opportunities:")
            for cat, amount in sorted_expenses[:3]:
                reduction = amount * 0.15  # Suggest 15% reduction
                suggestions.append(f"  ‚Ä¢ Reduce {cat.replace('_', ' ')}: Save ‚Çπ{reduction:,.0f}")

        elif analysis['savings_rate'] < 10:
            suggestions.append("‚ö†Ô∏è Low savings rate. Aim for at least 20% of income.")

        else:
            suggestions.append(f"‚úÖ Good job! Saving {analysis['savings_rate']:.1f}% of income")

        # Get AI insights
        ai_query = f"Based on my budget analysis, what are the top 3 specific actions I should take?"
        ai_suggestion = ai_advisor.analyze_with_ai(ai_query, analysis)
        suggestions.append(f"\nü§ñ AI Insights:\n{ai_suggestion}")

        return suggestions

# ============================================================================
# VISUALIZATION
# ============================================================================

class FinancialVisualizer:
    """Create visual dashboards"""

    @staticmethod
    def create_dashboard(analysis: Dict):
        """Create comprehensive financial dashboard"""

        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        fig.suptitle('üí∞ Financial Dashboard', fontsize=16, fontweight='bold')

        # 1. Income vs Expenses Bar Chart
        ax1 = axes[0, 0]
        categories_summary = ['Income', 'Expenses', 'Savings']
        values = [
            analysis['total_income'],
            analysis['total_expenses'],
            analysis['savings']
        ]
        colors = ['#2ecc71', '#e74c3c', '#3498db']
        ax1.bar(categories_summary, values, color=colors, alpha=0.7)
        ax1.set_title('Income vs Expenses vs Savings')
        ax1.set_ylabel('Amount (‚Çπ)')
        for i, v in enumerate(values):
            ax1.text(i, v, f'‚Çπ{v:,.0f}', ha='center', va='bottom')

        # 2. Expense Category Breakdown Pie Chart
        ax2 = axes[0, 1]
        if analysis['category_breakdown']:
            categories = list(analysis['category_breakdown'].keys())
            amounts = list(analysis['category_breakdown'].values())
            ax2.pie(amounts, labels=[c.replace('_', ' ').title() for c in categories],
                   autopct='%1.1f%%', startangle=90)
            ax2.set_title('Expense Distribution by Category')

        # 3. 50-30-20 Rule Comparison
        ax3 = axes[1, 0]
        rule_categories = ['Needs', 'Wants', 'Savings']
        recommended = [
            analysis['needs_vs_wants']['needs_budget'],
            analysis['needs_vs_wants']['wants_budget'],
            analysis['needs_vs_wants']['savings_budget']
        ]
        actual = [
            analysis['needs_vs_wants']['needs_actual'],
            analysis['needs_vs_wants']['wants_actual'],
            analysis['needs_vs_wants']['savings_actual']
        ]

        x = range(len(rule_categories))
        width = 0.35
        ax3.bar([i - width/2 for i in x], recommended, width, label='Recommended', alpha=0.7)
        ax3.bar([i + width/2 for i in x], actual, width, label='Actual', alpha=0.7)
        ax3.set_xlabel('Categories')
        ax3.set_ylabel('Amount (‚Çπ)')
        ax3.set_title('50-30-20 Rule: Recommended vs Actual')
        ax3.set_xticks(x)
        ax3.set_xticklabels(rule_categories)
        ax3.legend()

        # 4. Savings Rate Gauge
        ax4 = axes[1, 1]
        savings_rate = analysis['savings_rate']
        colors_gauge = ['#e74c3c' if savings_rate < 10 else '#f39c12' if savings_rate < 20 else '#2ecc71']
        ax4.barh(['Savings Rate'], [savings_rate], color=colors_gauge, alpha=0.7)
        ax4.set_xlim(0, 100)
        ax4.set_xlabel('Percentage (%)')
        ax4.set_title(f'Savings Rate: {savings_rate:.1f}%')
        ax4.text(savings_rate, 0, f'{savings_rate:.1f}%', va='center')

        # Add target line
        ax4.axvline(x=20, color='green', linestyle='--', label='Target: 20%')
        ax4.legend()

        plt.tight_layout()
        plt.show()

    @staticmethod
    def show_expense_trend(expenses: List[Dict]):
        """Show expense trends over time"""
        if not expenses:
            print("No expense data to visualize")
            return

        df = pd.DataFrame(expenses)
        df['date'] = pd.to_datetime(df['date'])
        daily_expenses = df.groupby('date')['amount'].sum()

        plt.figure(figsize=(12, 5))
        plt.plot(daily_expenses.index, daily_expenses.values, marker='o')
        plt.title('Daily Expense Trend')
        plt.xlabel('Date')
        plt.ylabel('Amount (‚Çπ)')
        plt.xticks(rotation=45)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

# ============================================================================
# GAMIFICATION SYSTEM
# ============================================================================

class GamificationEngine:
    """Gamification features for user engagement"""

    CHALLENGES = [
        {"name": "No Dining Out Week", "target": 0, "category": "entertainment",
         "duration": 7, "points": 500, "difficulty": "medium"},
        {"name": "Save ‚Çπ1000 This Week", "target": 1000, "type": "savings",
         "duration": 7, "points": 300, "difficulty": "easy"},
        {"name": "Reduce Utility Bills by 20%", "target": 0.20, "category": "utilities",
         "duration": 30, "points": 1000, "difficulty": "hard"},
        {"name": "Cook at Home 5 Days", "target": 5, "category": "food_groceries",
         "duration": 7, "points": 400, "difficulty": "medium"},
    ]

    def __init__(self):
        self.points = 0
        self.active_challenges = []
        self.completed_challenges = []
        self.badges = []

    def suggest_challenge(self, user_category: str, savings_rate: float) -> Dict:
        """Suggest appropriate challenge based on user profile"""

        if savings_rate < 10:
            # Suggest saving challenges
            return self.CHALLENGES[1]  # Save ‚Çπ1000
        else:
            # Suggest expense reduction challenges
            return self.CHALLENGES[0]  # No dining out

    def check_challenge_progress(self, expenses: List[Dict], challenge: Dict) -> Dict:
        """Check progress on active challenge"""

        # Implementation for progress tracking
        progress = {
            'challenge_name': challenge['name'],
            'progress_percent': 0,
            'status': 'in_progress',
            'points_earned': 0
        }

        return progress

    def award_points(self, points: int, reason: str):
        """Award points to user"""
        self.points += points
        print(f"\nüéâ Congratulations! You earned {points} points for: {reason}")
        print(f"üíé Total Points: {self.points}")

# ============================================================================
# MAIN APPLICATION
# ============================================================================

class FinancialAdvisorApp:
    """Main application orchestrator"""

    def __init__(self):
        """Initialize the application"""
        self.ai_advisor = GraniteFinancialAdvisor()
        self.profile_manager = UserProfileManager()
        self.tracker = FinancialTracker()
        self.analyzer = BudgetAnalyzer()
        self.visualizer = FinancialVisualizer()
        self.gamification = GamificationEngine()

        self.user_profile = None
        self.income_list = []
        self.expenses_list = []
        self.analysis = None

    def start(self):
        """Start the application"""
        print("\n" + "üåü" * 30)
        print("   AI-POWERED FINANCIAL ADVISOR")
        print("üåü" * 30)

        # Onboarding
        self.user_profile = self.profile_manager.conduct_onboarding()

        # Collect financial data
        self.income_list = self.tracker.collect_income(self.user_profile['category'])
        total_income = sum(inc['amount'] for inc in self.income_list)

        self.expenses_list = self.tracker.collect_expenses_interactive()

        # Analyze
        self.analysis = self.analyzer.analyze_budget(
            total_income,
            self.expenses_list,
            self.user_profile
        )

        # Show results
        self.show_analysis()

        # Interactive menu
        self.interactive_menu()

    def show_analysis(self):
        """Display analysis results"""
        print("\n" + "=" * 60)
        print("üìä FINANCIAL ANALYSIS REPORT")
        print("=" * 60)

        print(f"\nüí∞ Total Income: ‚Çπ{self.analysis['total_income']:,.0f}")
        print(f"üí∏ Total Expenses: ‚Çπ{self.analysis['total_expenses']:,.0f}")
        print(f"üíµ Savings: ‚Çπ{self.analysis['savings']:,.0f}")
        print(f"üìà Savings Rate: {self.analysis['savings_rate']:.1f}%")

        # Get AI suggestions
        suggestions = self.analyzer.suggest_adjustments(self.analysis, self.ai_advisor)
        print("\nüí° RECOMMENDATIONS:")
        for suggestion in suggestions:
            print(suggestion)

        # Visualize
        print("\nüìä Generating visual dashboard...")
        self.visualizer.create_dashboard(self.analysis)

    def interactive_menu(self):
        """Interactive menu for user queries"""
        print("\n" + "=" * 60)
        print("ü§ñ INTERACTIVE AI ASSISTANT")
        print("=" * 60)
        print("\nYou can now ask me anything about your finances!")
        print("Examples:")
        print("  - How can I save more money?")
        print("  - Which expenses should I cut?")
        print("  - Am I spending too much on entertainment?")
        print("  - Give me tax saving suggestions")
        print("  - Show me my spending pattern")
        print("\nType 'exit' to quit\n")

        while True:
            query = input("üí¨ You: ").strip()

            if query.lower() in ['exit', 'quit', 'bye']:
                print("\nüëã Thank you for using AI Financial Advisor!")
                break

            if not query:
                continue

            # Handle specific commands
            if 'visualize' in query.lower() or 'chart' in query.lower():
                self.visualizer.create_dashboard(self.analysis)
                continue

            if 'challenge' in query.lower():
                challenge = self.gamification.suggest_challenge(
                    self.user_profile['category'],
                    self.analysis['savings_rate']
                )
                print(f"\nüéØ Suggested Challenge: {challenge['name']}")
                print(f"   Duration: {challenge['duration']} days")
                print(f"   Reward: {challenge['points']} points")
                accept = input("Accept challenge? (yes/no): ").lower()
                if accept == 'yes':
                    self.gamification.active_challenges.append(challenge)
                    print("‚úÖ Challenge accepted! Good luck! üöÄ")
                continue

            # Get AI response
            print("\nü§ñ AI Advisor: ", end="")
            response = self.ai_advisor.analyze_with_ai(query, {
                'category': self.user_profile['category'],
                'budget': self.analysis['total_income'],
                'total_income': self.analysis['total_income'],
                'total_expenses': self.analysis['total_expenses'],
                'savings': self.analysis['savings'],
                'location': self.user_profile.get('city', 'Unknown'),
                'goals': self.user_profile.get('goals', [])
            })
            print(response)
            print()

# ============================================================================
# ADDITIONAL FEATURES
# ============================================================================

class AlertSystem:
    """Alert and notification system"""

    @staticmethod
    def check_bills_due(bills: List[Dict]) -> List[str]:
        """Check for upcoming bill due dates"""
        alerts = []
        today = datetime.now()

        for bill in bills:
            due_date = datetime.strptime(bill['due_date'], "%Y-%m-%d")
            days_until_due = (due_date - today).days

            if days_until_due <= 3 and days_until_due >= 0:
                alerts.append(f"‚ö†Ô∏è {bill['name']} due in {days_until_due} days (‚Çπ{bill['amount']:,.0f})")
            elif days_until_due < 0:
                alerts.append(f"üö® OVERDUE: {bill['name']} (‚Çπ{bill['amount']:,.0f})")

        return alerts

    @staticmethod
    def check_overspending(expenses: List[Dict], budget: float, category: str = None) -> List[str]:
        """Check for overspending alerts"""
        alerts = []

        if category:
            cat_expenses = sum(e['amount'] for e in expenses if e['category'] == category)
            cat_budget = budget * 0.15  # Assume 15% of budget per major category

            if cat_expenses > cat_budget:
                alerts.append(f"‚ö†Ô∏è Overspending on {category.replace('_', ' ')}: "
                            f"‚Çπ{cat_expenses:,.0f} / ‚Çπ{cat_budget:,.0f}")
        else:
            total_expenses = sum(e['amount'] for e in expenses)
            if total_expenses > budget:
                alerts.append(f"üö® BUDGET EXCEEDED by ‚Çπ{(total_expenses - budget):,.0f}")

        return alerts

class ComparativeAnalysis:
    """Compare user with others in similar category"""

    # Sample benchmark data (in real app, this would come from database)
    BENCHMARKS = {
        'student': {
            'avg_income': 8000,
            'avg_expenses': 6500,
            'avg_savings_rate': 18.75,
            'top_expenses': ['food_groceries', 'transportation', 'entertainment']
        },
        'employed': {
            'avg_income': 45000,
            'avg_expenses': 35000,
            'avg_savings_rate': 22.22,
            'top_expenses': ['rent', 'food_groceries', 'transportation']
        },
        'housewife': {
            'avg_income': 0,  # Dependent on family
            'avg_expenses': 25000,
            'avg_savings_rate': 15.0,
            'top_expenses': ['food_groceries', 'household_items', 'utilities']
        }
    }

    @staticmethod
    def compare_with_peers(user_data: Dict, category: str) -> Dict:
        """Compare user's financial behavior with peers"""

        benchmark = ComparativeAnalysis.BENCHMARKS.get(category,
                    ComparativeAnalysis.BENCHMARKS['employed'])

        income_diff = ((user_data['total_income'] - benchmark['avg_income']) /
                       benchmark['avg_income'] * 100) if benchmark['avg_income'] > 0 else 0

        expense_diff = ((user_data['total_expenses'] - benchmark['avg_expenses']) /
                        benchmark['avg_expenses'] * 100) if benchmark['avg_expenses'] > 0 else 0

        savings_diff = user_data['savings_rate'] - benchmark['avg_savings_rate']

        comparison = {
            'income_vs_peers': f"{income_diff:+.1f}%",
            'expense_vs_peers': f"{expense_diff:+.1f}%",
            'savings_vs_peers': f"{savings_diff:+.1f}%",
            'status': 'better' if savings_diff > 0 else 'needs_improvement',
            'message': ''
        }

        if savings_diff > 5:
            comparison['message'] = f"üéâ Excellent! You're saving {savings_diff:.1f}% more than peers!"
        elif savings_diff > 0:
            comparison['message'] = f"‚úÖ Good! You're saving slightly more than average."
        else:
            comparison['message'] = f"‚ö†Ô∏è You're saving {abs(savings_diff):.1f}% less than peers. Let's improve!"

        return comparison

class TaxAndInvestmentAdvisor:
    """Provide tax and investment suggestions"""

    TAX_SAVING_OPTIONS = {
        'Section 80C': {
            'limit': 150000,
            'options': ['PPF', 'ELSS', 'Life Insurance', 'NSC', 'Tuition Fees', '5-Year FD']
        },
        'Section 80D': {
            'limit': 25000,
            'options': ['Health Insurance Premium']
        },
        'Section 80G': {
            'limit': None,
            'options': ['Donations to Charitable Organizations']
        },
        'NPS (80CCD)': {
            'limit': 50000,
            'options': ['National Pension Scheme']
        }
    }

    @staticmethod
    def suggest_tax_savings(annual_income: float, user_profile: Dict) -> List[str]:
        """Suggest tax-saving investments"""
        suggestions = []

        suggestions.append("üí∞ TAX SAVING OPPORTUNITIES (Indian Tax Law):")
        suggestions.append("\nüìå Section 80C (Save up to ‚Çπ46,800 in taxes):")
        suggestions.append("   ‚Ä¢ PPF - Safe, government-backed (7.1% interest)")
        suggestions.append("   ‚Ä¢ ELSS Mutual Funds - Higher returns, 3-year lock-in")
        suggestions.append("   ‚Ä¢ Life Insurance - Protection + tax benefit")

        if user_profile.get('age', 30) < 40:
            suggestions.append("\nüìå For Young Investors:")
            suggestions.append("   ‚Ä¢ Start SIP in diversified mutual funds")
            suggestions.append("   ‚Ä¢ Consider Index Funds (low cost, market returns)")

        if user_profile.get('category') == 'employed':
            suggestions.append("\nüìå Section 80D (Health Insurance):")
            suggestions.append("   ‚Ä¢ Save up to ‚Çπ7,800 in taxes")
            suggestions.append("   ‚Ä¢ Essential for medical emergencies")

        if user_profile.get('age', 30) > 50:
            suggestions.append("\nüìå For Senior Citizens:")
            suggestions.append("   ‚Ä¢ Senior Citizen Savings Scheme (8.2% interest)")
            suggestions.append("   ‚Ä¢ Higher health insurance deduction (‚Çπ50,000)")

        return suggestions

    @staticmethod
    def suggest_investment_portfolio(risk_profile: str, monthly_savings: float) -> Dict:
        """Suggest investment portfolio based on risk profile"""

        portfolios = {
            'conservative': {
                'PPF': 40,
                'Fixed Deposits': 30,
                'Debt Mutual Funds': 20,
                'Gold': 10
            },
            'moderate': {
                'Equity Mutual Funds': 40,
                'Debt Mutual Funds': 30,
                'PPF': 20,
                'Gold': 10
            },
            'aggressive': {
                'Equity Mutual Funds': 60,
                'Equity Stocks': 20,
                'Debt Funds': 15,
                'Gold': 5
            }
        }

        portfolio = portfolios.get(risk_profile, portfolios['moderate'])

        allocation = {}
        for instrument, percentage in portfolio.items():
            allocation[instrument] = (monthly_savings * percentage / 100)

        return allocation

class GovernmentSchemes:
    """Information about government schemes and benefits"""

    SCHEMES = {
        'students': [
            {'name': 'PM Scholarship Scheme', 'benefit': '‚Çπ2,000-3,000/month',
             'eligibility': 'Merit-based for economically weaker sections'},
            {'name': 'NSP (National Scholarship Portal)', 'benefit': 'Various amounts',
             'eligibility': 'State and central scholarships'},
        ],
        'housewife': [
            {'name': 'PM Jan Dhan Yojana', 'benefit': 'Free bank account + insurance',
             'eligibility': 'All Indian citizens'},
            {'name': 'Mahila Samman Savings Certificate', 'benefit': '7.5% interest',
             'eligibility': 'Women investors'},
        ],
        'senior_citizen': [
            {'name': 'Senior Citizen Savings Scheme', 'benefit': '8.2% interest',
             'eligibility': 'Age 60+'},
            {'name': 'Pradhan Mantri Vaya Vandana Yojana', 'benefit': 'Guaranteed pension',
             'eligibility': 'Age 60+'},
        ],
        'unemployed': [
            {'name': 'PM Kaushal Vikas Yojana', 'benefit': 'Free skill training',
             'eligibility': 'Youth seeking employment'},
            {'name': 'Startup India', 'benefit': 'Tax benefits, funding support',
             'eligibility': 'Entrepreneurs'},
        ]
    }

    @staticmethod
    def get_applicable_schemes(user_profile: Dict) -> List[Dict]:
        """Get applicable government schemes"""

        category = user_profile.get('category', '')
        age = user_profile.get('age', 0)

        applicable = []

        if category in GovernmentSchemes.SCHEMES:
            applicable.extend(GovernmentSchemes.SCHEMES[category])

        if age >= 60:
            applicable.extend(GovernmentSchemes.SCHEMES.get('senior_citizen', []))

        # Universal schemes
        applicable.append({
            'name': 'PM Suraksha Bima Yojana',
            'benefit': '‚Çπ2 lakh accident insurance for ‚Çπ20/year',
            'eligibility': 'Age 18-70 with bank account'
        })

        return applicable

# ============================================================================
# FAMILY INTEGRATION MODULE
# ============================================================================

class FamilyFinanceManager:
    """Manage family-level financial integration"""

    def __init__(self):
        self.family_members = []
        self.consolidated_data = {}

    def add_family_member(self, member_profile: Dict):
        """Add a family member"""
        self.family_members.append(member_profile)
        print(f"‚úÖ Added {member_profile['name']} to family")

    def consolidate_finances(self) -> Dict:
        """Consolidate all family members' finances"""

        total_income = 0
        total_expenses = 0
        total_savings = 0
        earning_members = 0

        for member in self.family_members:
            if member.get('income', 0) > 0:
                earning_members += 1
                total_income += member.get('income', 0)

            total_expenses += member.get('expenses', 0)

        total_savings = total_income - total_expenses

        self.consolidated_data = {
            'total_income': total_income,
            'total_expenses': total_expenses,
            'total_savings': total_savings,
            'earning_members': earning_members,
            'total_members': len(self.family_members),
            'per_capita_income': total_income / len(self.family_members) if self.family_members else 0,
            'per_capita_expense': total_expenses / len(self.family_members) if self.family_members else 0
        }

        return self.consolidated_data

    def show_family_dashboard(self):
        """Display family-level financial dashboard"""

        if not self.consolidated_data:
            self.consolidate_finances()

        print("\n" + "=" * 60)
        print("üë®‚Äçüë©‚Äçüëß‚Äçüë¶ FAMILY FINANCIAL DASHBOARD")
        print("=" * 60)

        print(f"\nüìä Family Overview:")
        print(f"   Total Members: {self.consolidated_data['total_members']}")
        print(f"   Earning Members: {self.consolidated_data['earning_members']}")

        print(f"\nüí∞ Consolidated Finances:")
        print(f"   Total Family Income: ‚Çπ{self.consolidated_data['total_income']:,.0f}")
        print(f"   Total Family Expenses: ‚Çπ{self.consolidated_data['total_expenses']:,.0f}")
        print(f"   Total Family Savings: ‚Çπ{self.consolidated_data['total_savings']:,.0f}")

        print(f"\nüë§ Per Capita:")
        print(f"   Income per member: ‚Çπ{self.consolidated_data['per_capita_income']:,.0f}")
        print(f"   Expenses per member: ‚Çπ{self.consolidated_data['per_capita_expense']:,.0f}")

# ============================================================================
# DAILY TRACKING WITH SMART INPUT
# ============================================================================

class DailyTracker:
    """Smart daily expense tracking to avoid user fatigue"""

    def __init__(self):
        self.daily_log = []
        self.tracking_mode = 'smart'  # 'smart', 'quick', 'detailed'

    def quick_daily_check(self) -> Dict:
        """Quick daily expense check - minimize user effort"""

        print("\n" + "=" * 60)
        print("üì± QUICK DAILY CHECK (Takes 30 seconds!)")
        print("=" * 60)

        today = datetime.now().strftime("%Y-%m-%d")

        # Smart questions
        print("\nüí¨ Just answer these quick questions:")

        # Question 1: Did you spend money today?
        spent_today = input("Did you spend money today? (yes/no): ").lower() == 'yes'

        if not spent_today:
            print("‚úÖ Great job on a no-spend day! +50 points üéâ")
            return {'date': today, 'total': 0, 'no_spend_day': True}

        # Question 2: Quick estimate
        print("\nüí≠ Quick estimate - about how much did you spend today?")
        print("   1. Less than ‚Çπ200 (minimal)")
        print("   2. ‚Çπ200-500 (moderate)")
        print("   3. ‚Çπ500-1000 (average)")
        print("   4. More than ‚Çπ1000 (high)")

        choice = int(input("Pick (1-4): "))

        ranges = {1: 100, 2: 350, 3: 750, 4: 1500}
        estimated_amount = ranges.get(choice, 500)

        # Question 3: Main category
        print("\nüè∑Ô∏è What did you mainly spend on?")
        quick_categories = ['Food', 'Transport', 'Shopping', 'Bills', 'Entertainment', 'Other']
        for i, cat in enumerate(quick_categories, 1):
            print(f"   {i}. {cat}")

        cat_choice = int(input("Pick (1-6): "))
        category = quick_categories[cat_choice - 1].lower()

        expense_entry = {
            'date': today,
            'amount': estimated_amount,
            'category': category,
            'method': 'quick_estimate',
            'time': datetime.now().strftime("%H:%M:%S")
        }

        self.daily_log.append(expense_entry)

        print(f"\n‚úÖ Logged ‚Çπ{estimated_amount:,.0f} for {category}")
        print("   (You can add detailed expenses anytime later)")

        return expense_entry

    def weekly_summary_prompt(self, expenses: List[Dict]) -> str:
        """Generate weekly summary to keep user engaged"""

        week_total = sum(e['amount'] for e in expenses[-7:])

        summary = f"""

üìä YOUR WEEK IN REVIEW
{'='*60}

This Week's Spending: ‚Çπ{week_total:,.0f}

üéØ Keep it up! You're tracking your expenses consistently.
üí° Tip: Review your weekly spending every Sunday to stay on track.

Want to see detailed breakdown? Just ask me!
        """

        return summary

# ============================================================================
# RUN THE APPLICATION
# ============================================================================

def main():
    """Main entry point"""

    try:
        # Initialize and start the app
        app = FinancialAdvisorApp()
        app.start()

        # Additional features menu
        print("\n" + "=" * 60)
        print("üéÅ ADDITIONAL FEATURES")
        print("=" * 60)
        print("\n1. View Government Schemes")
        print("2. Get Tax Saving Suggestions")
        print("3. Compare with Peers")
        print("4. Set Up Daily Tracking")
        print("5. View Alerts")
        print("6. Export Report")
        print("0. Continue to chat")

        choice = input("\nSelect an option (0-6): ").strip()

        if choice == '1':
            schemes = GovernmentSchemes.get_applicable_schemes(app.user_profile)
            print("\nüíº APPLICABLE GOVERNMENT SCHEMES:")
            for scheme in schemes:
                print(f"\nüìã {scheme['name']}")
                print(f"   Benefit: {scheme['benefit']}")
                print(f"   Eligibility: {scheme['eligibility']}")

        elif choice == '2':
            annual_income = app.analysis['total_income'] * 12
            suggestions = TaxAndInvestmentAdvisor.suggest_tax_savings(
                annual_income,
                app.user_profile
            )
            print("\n")
            for suggestion in suggestions:
                print(suggestion)

        elif choice == '3':
            comparison = ComparativeAnalysis.compare_with_peers(
                app.analysis,
                app.user_profile['category']
            )
            print("\nüìä PEER COMPARISON:")
            print(f"\n{comparison['message']}")
            print(f"\nIncome vs Peers: {comparison['income_vs_peers']}")
            print(f"Expenses vs Peers: {comparison['expense_vs_peers']}")
            print(f"Savings Rate vs Peers: {comparison['savings_vs_peers']}")

        elif choice == '4':
            tracker = DailyTracker()
            tracker.quick_daily_check()

        elif choice == '5':
            alerts = AlertSystem.check_overspending(
                app.expenses_list,
                app.analysis['total_income']
            )
            if alerts:
                print("\n‚ö†Ô∏è ALERTS:")
                for alert in alerts:
                    print(f"   {alert}")
            else:
                print("\n‚úÖ No alerts! You're doing great!")

        print("\n‚ú® Thank you for using AI Financial Advisor!")
        print("üí° Remember: Small steps lead to big savings!")

    except KeyboardInterrupt:
        print("\n\nüëã Session interrupted. Your progress is saved!")
    except Exception as e:
        print(f"\n‚ùå An error occurred: {e}")
        print("üí° Please check your setup and try again.")

# ============================================================================
# ENTRY POINT
# ============================================================================

if __name__ == "__main__":
    print("""
    ‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó
    ‚ïë                                                              ‚ïë
    ‚ïë          ü§ñ AI FINANCIAL ADVISOR PROTOTYPE ü§ñ               ‚ïë
    ‚ïë                                                              ‚ïë
    ‚ïë     Powered by IBM Granite 3.2-2B Instruct Model            ‚ïë
    ‚ïë                                                              ‚ïë
    ‚ïë  Features:                                                   ‚ïë
    ‚ïë  ‚úì Smart onboarding & profiling                            ‚ïë
    ‚ïë  ‚úì AI-powered budget analysis                              ‚ïë
    ‚ïë  ‚úì Expense tracking & predictions                          ‚ïë
    ‚ïë  ‚úì Visual dashboards                                       ‚ïë
    ‚ïë  ‚úì Tax & investment suggestions                            ‚ïë
    ‚ïë  ‚úì Government schemes info                                 ‚ïë
    ‚ïë  ‚úì Peer comparison                                         ‚ïë
    ‚ïë  ‚úì Gamification & challenges                               ‚ïë
    ‚ïë  ‚úì Family finance integration                              ‚ïë
    ‚ïë                                                              ‚ïë
    ‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù
    """)

    print("\n‚öôÔ∏è  SETUP INSTRUCTIONS:")
    print("=" * 60)
    print("1. Make sure you uncommented the pip install line at the top")
    print("2. Run that cell first to install dependencies")
    print("3. Then run this cell to start the application")
    print("4. Follow the interactive prompts")
    print("=" * 60)

    input("\nPress Enter to continue...")

    main()
```

```python
"""
AI-Driven Financial Management Chatbot with Gradio UI
Uses IBM Granite 3.2-2B Instruct Model
Designed for Google Colab with Beautiful Interactive Interface

INSTALLATION - Run this first:
!pip install -q transformers torch accelerate huggingface_hub pandas matplotlib seaborn plotly gradio
"""

import gradio as gr
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
import json
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from typing import Dict, List, Optional, Tuple
import warnings
import io
import base64
from PIL import Image
warnings.filterwarnings('ignore')

# ============================================================================
# GLOBAL STATE MANAGEMENT
# ============================================================================

class AppState:
    """Global application state"""
    def __init__(self):
        self.model = None
        self.tokenizer = None
        self.user_profile = {}
        self.income_list = []
        self.expenses_list = []
        self.analysis = {}
        self.conversation_history = []
        self.onboarding_step = 0
        self.gamification_points = 0
        self.challenges = []
        self.is_model_loaded = False

app_state = AppState()

# ============================================================================
# MODEL LOADING
# ============================================================================

def load_model():
    """Load IBM Granite model"""
    if app_state.is_model_loaded:
        return "‚úÖ Model already loaded!"

    try:
        yield "üîÑ Loading IBM Granite 3.2-2B model... (2-3 minutes)"

        model_name = "ibm-granite/granite-3.2-2b-instruct"

        app_state.tokenizer = AutoTokenizer.from_pretrained(model_name)
        yield "‚úÖ Tokenizer loaded! Loading model..."

        app_state.model = AutoModelForCausalLM.from_pretrained(
            model_name,
            device_map="auto",
            torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
            low_cpu_mem_usage=True
        )

        app_state.is_model_loaded = True
        device = "GPU" if torch.cuda.is_available() else "CPU"
        yield f"‚úÖ Model loaded successfully on {device}!\n\nüéâ Ready to help you manage your finances!"

    except Exception as e:
        yield f"‚ùå Error loading model: {str(e)}\n\nPlease check your internet connection and try again."

def generate_ai_response(prompt: str, context: str = "") -> str:
    """Generate AI response"""
    if not app_state.is_model_loaded:
        return "‚ö†Ô∏è Please load the model first using the 'Load AI Model' button in Setup tab."

    if context:
        full_prompt = f"{context}\n\nUser: {prompt}\n\nFinancial Advisor:"
    else:
        full_prompt = f"You are a helpful financial advisor. Answer concisely:\n\n{prompt}\n\nAnswer:"

    inputs = app_state.tokenizer(full_prompt, return_tensors="pt", truncation=True, max_length=1024)
    inputs = {k: v.to(app_state.model.device) for k, v in inputs.items()}

    with torch.no_grad():
        outputs = app_state.model.generate(
            **inputs,
            max_new_tokens=250,
            temperature=0.7,
            top_p=0.9,
            do_sample=True,
            pad_token_id=app_state.tokenizer.eos_token_id
        )

    response = app_state.tokenizer.decode(outputs[0], skip_special_tokens=True)
    response = response.split("Answer:")[-1].strip()
    if "User:" in response:
        response = response.split("User:")[0].strip()

    return response

# ============================================================================
# ONBOARDING FUNCTIONS
# ============================================================================

def save_basic_info(name, age, category, marital_status):
    """Save basic user information"""
    if not name or not age:
        return "‚ö†Ô∏è Please fill in all fields", None

    app_state.user_profile.update({
        'name': name,
        'age': int(age),
        'category': category,
        'marital_status': marital_status,
        'created_at': datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    })

    summary = f"""‚úÖ **Profile Created!**

üë§ **Name:** {name}
üéÇ **Age:** {age}
üìä **Category:** {category}
üíë **Marital Status:** {marital_status}

Please continue to the next section!"""

    return summary, gr.update(visible=True)

def save_family_location(family_size, dependents, city, country, living_situation):
    """Save family and location info"""
    app_state.user_profile.update({
        'family_size': int(family_size),
        'dependents': int(dependents),
        'city': city,
        'country': country,
        'living_situation': living_situation
    })

    return f"‚úÖ Family and location info saved! Total family: {family_size}, Location: {city}"

def save_assets(owns_vehicle, vehicle_type, owns_property):
    """Save assets information"""
    app_state.user_profile.update({
        'owns_vehicle': owns_vehicle,
        'vehicle_type': vehicle_type if owns_vehicle else 'None',
        'owns_property': owns_property
    })

    return "‚úÖ Assets information saved!"

def save_goals(goals_text):
    """Save financial goals"""
    goals = [g.strip() for g in goals_text.split('\n') if g.strip()]
    app_state.user_profile['goals'] = goals

    profile_summary = f"""
üéâ **Profile Complete!**

**Summary:**
- Name: {app_state.user_profile.get('name')}
- Age: {app_state.user_profile.get('age')}
- Category: {app_state.user_profile.get('category')}
- Family Size: {app_state.user_profile.get('family_size')}
- Location: {app_state.user_profile.get('city')}, {app_state.user_profile.get('country')}
- Goals: {len(goals)} goals set

**Next Step:** Add your income sources in the Income tab!
"""

    return profile_summary

# ============================================================================
# INCOME MANAGEMENT
# ============================================================================

def add_income(source, amount, frequency):
    """Add income entry"""
    if not source or not amount:
        return "‚ö†Ô∏è Please fill in all fields", create_income_table()

    try:
        income_entry = {
            'source': source,
            'amount': float(amount),
            'frequency': frequency,
            'date': datetime.now().strftime("%Y-%m-%d")
        }

        app_state.income_list.append(income_entry)

        total = sum(inc['amount'] for inc in app_state.income_list)
        return f"‚úÖ Added ‚Çπ{amount} from {source}. Total income: ‚Çπ{total:,.0f}", create_income_table()

    except ValueError:
        return "‚ö†Ô∏è Invalid amount", create_income_table()

def create_income_table():
    """Create income summary table"""
    if not app_state.income_list:
        return pd.DataFrame(columns=['Source', 'Amount (‚Çπ)', 'Frequency'])

    df = pd.DataFrame(app_state.income_list)
    df_display = pd.DataFrame({
        'Source': df['source'],
        'Amount (‚Çπ)': df['amount'].apply(lambda x: f"‚Çπ{x:,.0f}"),
        'Frequency': df['frequency']
    })

    return df_display

def get_income_sources(category):
    """Get income sources based on category"""
    sources = {
        'Student': ['Pocket Money', 'Scholarship', 'Part-time Job', 'Tuition', 'Freelancing'],
        'Employed': ['Salary', 'Bonus', 'Investment Returns', 'Rental Income', 'Side Business'],
        'Unemployed': ['Savings', 'Family Support', 'Freelancing', 'Unemployment Benefits'],
        'Housewife': ['Family Income', 'Side Business', 'Investment Returns'],
        'Retired': ['Pension', 'Investment Returns', 'Rental Income', 'Savings']
    }
    return gr.update(choices=sources.get(category, sources['Employed']))

# ============================================================================
# EXPENSE MANAGEMENT
# ============================================================================

def add_expense_quick(description):
    """Quick expense entry with smart parsing"""
    if not description:
        return "‚ö†Ô∏è Please enter an expense", create_expense_table(), None

    try:
        # Parse: "description amount" or just "amount"
        parts = description.strip().split()
        amount = float(parts[-1])
        desc = ' '.join(parts[:-1]) if len(parts) > 1 else 'Expense'

        # Auto-categorize
        category = auto_categorize_expense(desc)

        expense_entry = {
            'description': desc,
            'amount': amount,
            'category': category,
            'date': datetime.now().strftime("%Y-%m-%d"),
            'time': datetime.now().strftime("%H:%M:%S")
        }

        app_state.expenses_list.append(expense_entry)

        total = sum(exp['amount'] for exp in app_state.expenses_list)
        return (
            f"‚úÖ Added ‚Çπ{amount:,.0f} - {desc} ({category})\nTotal expenses: ‚Çπ{total:,.0f}",
            create_expense_table(),
            ""  # Clear input
        )

    except (ValueError, IndexError):
        return "‚ö†Ô∏è Format: 'description amount' (e.g., 'Groceries 500')", create_expense_table(), description

def auto_categorize_expense(description: str) -> str:
    """Auto-categorize based on keywords"""
    description = description.lower()

    keywords = {
        'Food & Groceries': ['grocery', 'food', 'vegetables', 'supermarket', 'lunch', 'dinner', 'breakfast'],
        'Rent': ['rent', 'lease'],
        'Utilities': ['electricity', 'water', 'gas', 'internet', 'phone', 'mobile'],
        'Transportation': ['petrol', 'diesel', 'fuel', 'bus', 'taxi', 'uber', 'ola', 'auto'],
        'Healthcare': ['medicine', 'doctor', 'hospital', 'health', 'medical'],
        'Education': ['school', 'college', 'tuition', 'books', 'course', 'fees'],
        'Entertainment': ['movie', 'restaurant', 'party', 'gaming', 'netflix', 'amazon prime'],
        'Clothing': ['clothes', 'shoes', 'dress', 'shirt'],
    }

    for category, words in keywords.items():
        if any(word in description for word in words):
            return category

    return 'Other'

def add_expense_detailed(description, amount, category):
    """Detailed expense entry"""
    if not description or not amount:
        return "‚ö†Ô∏è Please fill in all fields", create_expense_table()

    try:
        expense_entry = {
            'description': description,
            'amount': float(amount),
            'category': category,
            'date': datetime.now().strftime("%Y-%m-%d"),
            'time': datetime.now().strftime("%H:%M:%S")
        }

        app_state.expenses_list.append(expense_entry)

        total = sum(exp['amount'] for exp in app_state.expenses_list)
        return f"‚úÖ Added ‚Çπ{amount} - {description}. Total: ‚Çπ{total:,.0f}", create_expense_table()

    except ValueError:
        return "‚ö†Ô∏è Invalid amount", create_expense_table()

def create_expense_table():
    """Create expense summary table"""
    if not app_state.expenses_list:
        return pd.DataFrame(columns=['Date', 'Description', 'Category', 'Amount (‚Çπ)'])

    df = pd.DataFrame(app_state.expenses_list)
    df_display = pd.DataFrame({
        'Date': df['date'],
        'Description': df['description'],
        'Category': df['category'],
        'Amount (‚Çπ)': df['amount'].apply(lambda x: f"‚Çπ{x:,.0f}")
    })

    return df_display.tail(10)  # Show last 10

# ============================================================================
# ANALYSIS & VISUALIZATION
# ============================================================================

def analyze_budget():
    """Analyze budget and generate insights"""
    if not app_state.income_list or not app_state.expenses_list:
        return "‚ö†Ô∏è Please add income and expenses first", None, None, None

    total_income = sum(inc['amount'] for inc in app_state.income_list)
    total_expenses = sum(exp['amount'] for exp in app_state.expenses_list)
    savings = total_income - total_expenses
    savings_rate = (savings / total_income * 100) if total_income > 0 else 0

    # Category breakdown
    category_totals = {}
    for exp in app_state.expenses_list:
        cat = exp['category']
        category_totals[cat] = category_totals.get(cat, 0) + exp['amount']

    app_state.analysis = {
        'total_income': total_income,
        'total_expenses': total_expenses,
        'savings': savings,
        'savings_rate': savings_rate,
        'category_breakdown': category_totals,
        'status': 'surplus' if savings > 0 else 'deficit'
    }

    # Generate summary
    summary = f"""
# üìä Financial Analysis Report

## üí∞ Overview
- **Total Income:** ‚Çπ{total_income:,.0f}
- **Total Expenses:** ‚Çπ{total_expenses:,.0f}
- **Savings:** ‚Çπ{savings:,.0f}
- **Savings Rate:** {savings_rate:.1f}%

## üìà Status
"""

    if savings > 0:
        summary += f"‚úÖ **Surplus:** You're saving ‚Çπ{savings:,.0f} per month!\n\n"
    else:
        summary += f"‚ö†Ô∏è **Deficit:** You're overspending by ‚Çπ{abs(savings):,.0f}\n\n"

    # AI Insights
    if app_state.is_model_loaded:
        summary += "## ü§ñ AI Insights\n\n"
        context = f"""User Profile:
- Category: {app_state.user_profile.get('category', 'Unknown')}
- Monthly Income: ‚Çπ{total_income:,.0f}
- Monthly Expenses: ‚Çπ{total_expenses:,.0f}
- Savings: ‚Çπ{savings:,.0f}
- Savings Rate: {savings_rate:.1f}%

Provide 3 specific, actionable tips to improve their finances."""

        ai_response = generate_ai_response("What should I do to improve my financial situation?", context)
        summary += ai_response

    # Create visualizations
    fig1 = create_income_vs_expense_chart()
    fig2 = create_expense_breakdown_chart()
    fig3 = create_savings_rate_gauge()

    return summary, fig1, fig2, fig3

def create_income_vs_expense_chart():
    """Create income vs expense bar chart"""
    if not app_state.analysis:
        return None

    fig = go.Figure(data=[
        go.Bar(name='Amount', x=['Income', 'Expenses', 'Savings'],
               y=[app_state.analysis['total_income'],
                  app_state.analysis['total_expenses'],
                  app_state.analysis['savings']],
               marker_color=['#2ecc71', '#e74c3c', '#3498db'])
    ])

    fig.update_layout(
        title='Income vs Expenses vs Savings',
        yaxis_title='Amount (‚Çπ)',
        height=400,
        showlegend=False
    )

    return fig

def create_expense_breakdown_chart():
    """Create expense breakdown pie chart"""
    if not app_state.analysis or not app_state.analysis.get('category_breakdown'):
        return None

    categories = list(app_state.analysis['category_breakdown'].keys())
    amounts = list(app_state.analysis['category_breakdown'].values())

    fig = go.Figure(data=[go.Pie(labels=categories, values=amounts, hole=0.3)])

    fig.update_layout(
        title='Expense Distribution by Category',
        height=400
    )

    return fig

def create_savings_rate_gauge():
    """Create savings rate gauge"""
    if not app_state.analysis:
        return None

    savings_rate = app_state.analysis['savings_rate']

    fig = go.Figure(go.Indicator(
        mode="gauge+number+delta",
        value=savings_rate,
        domain={'x': [0, 1], 'y': [0, 1]},
        title={'text': "Savings Rate (%)"},
        delta={'reference': 20},
        gauge={
            'axis': {'range': [None, 50]},
            'bar': {'color': "darkblue"},
            'steps': [
                {'range': [0, 10], 'color': "#ffcccb"},
                {'range': [10, 20], 'color': "#ffffcc"},
                {'range': [20, 50], 'color': "#90EE90"}
            ],
            'threshold': {
                'line': {'color': "red", 'width': 4},
                'thickness': 0.75,
                'value': 20
            }
        }
    ))

    fig.update_layout(height=400)

    return fig

# ============================================================================
# CHATBOT
# ============================================================================

def chatbot_response(message, history):
    """Chatbot conversation"""
    if not app_state.is_model_loaded:
        return "‚ö†Ô∏è Please load the AI model first from the Setup tab."

    if not app_state.user_profile:
        return "‚ö†Ô∏è Please complete your profile in the Onboarding tab first."

    # Build context
    context = f"""User Profile:
- Name: {app_state.user_profile.get('name')}
- Category: {app_state.user_profile.get('category')}
- Age: {app_state.user_profile.get('age')}
"""

    if app_state.analysis:
        context += f"""
Financial Summary:
- Monthly Income: ‚Çπ{app_state.analysis['total_income']:,.0f}
- Monthly Expenses: ‚Çπ{app_state.analysis['total_expenses']:,.0f}
- Savings: ‚Çπ{app_state.analysis['savings']:,.0f}
- Savings Rate: {app_state.analysis['savings_rate']:.1f}%
"""

    response = generate_ai_response(message, context)

    return response

# ============================================================================
# ADDITIONAL FEATURES
# ============================================================================

def get_tax_suggestions():
    """Get tax saving suggestions"""
    if not app_state.analysis:
        return "‚ö†Ô∏è Please complete budget analysis first"

    annual_income = app_state.analysis['total_income'] * 12

    suggestions = f"""
# üí∞ Tax Saving Opportunities

## Section 80C (Save up to ‚Çπ46,800 in taxes)
Maximum deduction: ‚Çπ1,50,000

**Best Options:**
- **PPF (Public Provident Fund):** Safe, government-backed, 7.1% interest
- **ELSS Mutual Funds:** Higher returns, 3-year lock-in, equity exposure
- **Life Insurance Premium:** Protection + tax benefit
- **NSC (National Savings Certificate):** Fixed returns, 5-year tenure

## Section 80D (Health Insurance)
Maximum deduction: ‚Çπ25,000 (‚Çπ50,000 for senior citizens)

**Benefits:**
- Covers health insurance premium for self, family, and parents
- Includes preventive health check-ups (‚Çπ5,000)

## NPS (National Pension Scheme) - Section 80CCD(1B)
Additional deduction: ‚Çπ50,000

**Benefits:**
- Long-term retirement planning
- Tax benefits + market-linked returns

## Estimated Annual Tax Savings
Based on your income of ‚Çπ{annual_income:,.0f}:

- Max saving through 80C: ‚Çπ46,800
- Max saving through 80D: ‚Çπ7,800
- Max saving through NPS: ‚Çπ15,600

**Total Potential Savings: ‚Çπ70,200 per year!**

---

üí° **Recommendation:** Start with PPF (‚Çπ12,500/month) and term insurance for immediate benefits!
"""

    return suggestions

def get_government_schemes():
    """Get applicable government schemes"""
    if not app_state.user_profile:
        return "‚ö†Ô∏è Please complete profile first"

    category = app_state.user_profile.get('category', '').lower()
    age = app_state.user_profile.get('age', 0)

    schemes = f"""
# üíº Applicable Government Schemes

"""

    if 'student' in category:
        schemes += """
## For Students

### 1. PM Scholarship Scheme
- **Benefit:** ‚Çπ2,000-3,000/month
- **Eligibility:** Merit-based for economically weaker sections
- **Apply:** scholarships.gov.in

### 2. National Scholarship Portal (NSP)
- **Benefit:** Various state and central scholarships
- **Eligibility:** Based on category and merit
- **Apply:** scholarships.gov.in

### 3. Post Matric Scholarship
- **Benefit:** Full tuition + maintenance
- **Eligibility:** SC/ST/OBC students
"""

    elif 'employed' in category:
        schemes += """
## For Employed

### 1. PM Suraksha Bima Yojana
- **Benefit:** ‚Çπ2 lakh accident insurance
- **Premium:** Only ‚Çπ20/year
- **Eligibility:** Age 18-70 with bank account

### 2. Atal Pension Yojana
- **Benefit:** Guaranteed pension ‚Çπ1,000-5,000/month
- **Eligibility:** Age 18-40
"""

    elif 'housewife' in category:
        schemes += """
## For Housewives

### 1. Mahila Samman Savings Certificate
- **Benefit:** 7.5% interest (better than FD)
- **Eligibility:** All women
- **Tenure:** 2 years

### 2. PM Jan Dhan Yojana
- **Benefit:** Free bank account + ‚Çπ2 lakh insurance
- **Eligibility:** All Indian citizens
"""

    if age >= 60:
        schemes += """
## For Senior Citizens

### 1. Senior Citizen Savings Scheme (SCSS)
- **Benefit:** 8.2% interest
- **Maximum:** ‚Çπ30 lakh
- **Eligibility:** Age 60+

### 2. Pradhan Mantri Vaya Vandana Yojana
- **Benefit:** Guaranteed pension
- **Returns:** 7.4% per annum
"""

    schemes += """
---

## Universal Schemes

### PM Suraksha Bima Yojana
- ‚Çπ2 lakh accident cover for ‚Çπ20/year
- Age 18-70

### PM Jeevan Jyoti Bima Yojana
- ‚Çπ2 lakh life cover for ‚Çπ436/year
- Age 18-50

---

üí° Visit your nearest bank or post office to apply!
"""

    return schemes

def compare_with_peers():
    """Compare with peer benchmarks"""
    if not app_state.analysis:
        return "‚ö†Ô∏è Please complete budget analysis first"

    category = app_state.user_profile.get('category', 'employed').lower()

    benchmarks = {
        'student': {'avg_income': 8000, 'avg_expenses': 6500, 'avg_savings_rate': 18.75},
        'employed': {'avg_income': 45000, 'avg_expenses': 35000, 'avg_savings_rate': 22.22},
        'housewife': {'avg_income': 0, 'avg_expenses': 25000, 'avg_savings_rate': 15.0},
        'retired': {'avg_income': 30000, 'avg_expenses': 22000, 'avg_savings_rate': 26.67}
    }

    benchmark = benchmarks.get(category, benchmarks['employed'])

    user_savings_rate = app_state.analysis['savings_rate']
    savings_diff = user_savings_rate - benchmark['avg_savings_rate']

    comparison = f"""
# üìä Peer Comparison

## Your Category: {category.title()}

### Savings Rate Comparison
- **Your Savings Rate:** {user_savings_rate:.1f}%
- **Peer Average:** {benchmark['avg_savings_rate']:.1f}%
- **Difference:** {savings_diff:+.1f}%

"""

    if savings_diff > 5:
        comparison += f"""
üéâ **Excellent!** You're saving {savings_diff:.1f}% more than your peers!

You're in the top 20% of savers in your category. Keep it up!
"""
    elif savings_diff > 0:
        comparison += f"""
‚úÖ **Good Job!** You're saving more than average.

You're doing better than most people in your category. Small improvements can put you in the top tier!
"""
    else:
        comparison += f"""
‚ö†Ô∏è **Opportunity to Improve**

You're saving {abs(savings_diff):.1f}% less than peers in your category.

**Action Plan:**
1. Review your expense categories
2. Identify non-essential spending
3. Set a goal to increase savings by 5% this month
"""

    comparison += f"""

### Income vs Expenses
- **Peer Average Income:** ‚Çπ{benchmark['avg_income']:,.0f}
- **Peer Average Expenses:** ‚Çπ{benchmark['avg_expenses']:,.0f}

---

üí° **Tip:** Focus on savings rate rather than absolute amounts. Even small percentages compound over time!
"""

    return comparison

# ============================================================================
# GRADIO INTERFACE
# ============================================================================

def create_ui():
    """Create Gradio interface"""

    with gr.Blocks(title="AI Financial Advisor", theme=gr.themes.Soft()) as app:

        gr.Markdown("""
        # ü§ñ AI Financial Advisor
        ### Powered by IBM Granite 3.2-2B | Your Personal Finance Assistant
        """)

        with gr.Tabs() as tabs:

            # ============================================================
            # TAB 1: SETUP
            # ============================================================
            with gr.Tab("üöÄ Setup", id=0):
                gr.Markdown("""
                ## Welcome! Let's get started

                **Step 1:** Load the AI model (takes 2-3 minutes first time)
                **Step 2:** Complete your profile in the Onboarding tab
                **Step 3:** Add your income and expenses
                **Step 4:** Get AI-powered insights!
                """)

                load_btn = gr.Button("üîÑ Load AI Model", variant="primary", size="lg")
                model_status = gr.Textbox(label="Status", lines=3)

                load_btn.click(
                    fn=load_model,
                    outputs=[model_status]
                )

                gr.Markdown("""
                ---
                ### Quick Start Guide

                1. **Load Model:** Click the button above (one-time setup)
                2. **Profile:** Fill your details in Onboarding tab
                3. **Income:** Add all income sources
                4. **Expenses:** Track your spending (quick or detailed)
                5. **Analysis:** Get AI insights and visualizations
                6. **Chat:** Ask questions anytime!
                """)

            # ============================================================
            # TAB 2: ONBOARDING
            # ============================================================
            with gr.Tab("üë§ Onboarding", id=1):
                gr.Markdown("## Create Your Financial Profile")

                with gr.Group():
                    gr.Markdown("### Basic Information")
                    with gr.Row():
                        name_input = gr.Textbox(label="Name", placeholder="Enter your name")
                        age_input = gr.Number(label="Age", value=25, minimum=1, maximum=100)

                    with gr.Row():
                        category_input = gr.Dropdown(
                            label="Category",
                            choices=["Student", "Employed", "Unemployed", "Housewife", "Retired"],
                            value="Employed"
                        )
                        marital_input = gr.Dropdown(
                            label="Marital Status",
                            choices=["Single", "Married", "Divorced", "Widowed"],
                            value="Single"
                        )

                    basic_submit = gr.Button("Save Basic Info", variant="primary")
                    basic_output = gr.Markdown()

                with gr.Group(visible=True) as family_section:
                    gr.Markdown("### Family & Location")
                    with gr.Row():
                        family_size_input = gr.Number(label="Total Family Members", value=1, minimum=1)
                        dependents_input = gr.Number(label="Dependents", value=0, minimum=0)

                    with gr.Row():
                        city_input = gr.Textbox(label="City", placeholder="e.g., Bangalore")
                        country_input = gr.Textbox(label="Country", value="India")

                    living_input = gr.Dropdown(
                        label="Living Situation",
                        choices=["Own Home", "Rented", "PG", "Hostel", "With Parents"],
                        value="Rented"
                    )

                    family_submit = gr.Button("Save Family Info")
                    family_output = gr.Markdown()

                with gr.Group():
                    gr.Markdown("### Assets")
                    owns_vehicle = gr.Checkbox(label="Own Vehicle?")
                    vehicle_type = gr.Dropdown(
                        label="Vehicle Type",
                        choices=["Car", "Bike", "Both", "None"],
                        value="None"
                    )
                    owns_property = gr.Checkbox(label="Own Property?")

                    assets_submit = gr.Button("Save Assets")
                    assets_output = gr.Markdown()

                with gr.Group():
                    gr.Markdown("### Financial Goals & Dreams")
                    goals_input = gr.Textbox(
                        label="Your Goals (one per line)",
                        placeholder="Example:\nBuy a car in 2 years\nSave for education\nGo on Europe trip",
                        lines=5
                    )

                    goals_submit = gr.Button("Complete Profile", variant="primary")
                    profile_summary = gr.Markdown()

                # Connect buttons
                basic_submit.click(
                    fn=save_basic_info,
                    inputs=[name_input, age_input, category_input, marital_input],
                    outputs=[basic_output, family_section]
                )

                family_submit.click(
                    fn=save_family_location,
                    inputs=[family_size_input, dependents_input, city_input, country_input, living_input],
                    outputs=[family_output]
                )

                assets_submit.click(
                    fn=save_assets,
                    inputs=[owns_vehicle, vehicle_type, owns_property],
                    outputs=[assets_output]
                )

                goals_submit.click(
                    fn=save_goals,
                    inputs=[goals_input],
                    outputs=[profile_summary]
                )

            # ============================================================
            # TAB 3: INCOME
            # ============================================================
            with gr.Tab("üí∞ Income", id=2):
                gr.Markdown("## Add Your Income Sources")

                with gr.Row():
                    with gr.Column(scale=2):
                        income_source = gr.Dropdown(
                            label="Income Source",
                            choices=["Salary", "Bonus", "Investment Returns", "Rental Income", "Side Business", "Freelancing", "Other"],
                            value="Salary"
                        )
                        income_amount = gr.Number(label="Monthly Amount (‚Çπ)", value=0, minimum=0)
                        income_frequency = gr.Dropdown(
                            label="Frequency",
                            choices=["Monthly", "Yearly", "One-time"],
                            value="Monthly"
                        )

                        add_income_btn = gr.Button("‚ûï Add Income", variant="primary")
                        income_output = gr.Markdown()

                    with gr.Column(scale=3):
                        income_table = gr.Dataframe(
                            label="Your Income Sources",
                            headers=["Source", "Amount (‚Çπ)", "Frequency"],
                            interactive=False
                        )

                # Update income sources based on category
                category_input.change(
                    fn=get_income_sources,
                    inputs=[category_input],
                    outputs=[income_source]
                )

                add_income_btn.click(
                    fn=add_income,
                    inputs=[income_source, income_amount, income_frequency],
                    outputs=[income_output, income_table]
                )

                gr.Markdown("""
                ---
                ### üí° Tips for Income Entry
                - Include all regular income sources
                - For students: pocket money, scholarships, part-time work
                - For employed: salary, bonuses, side income
                - For investments: average monthly returns
                """)

            # ============================================================
            # TAB 4: EXPENSES
            # ============================================================
            with gr.Tab("üí∏ Expenses", id=3):
                gr.Markdown("## Track Your Expenses")

                with gr.Tabs():
                    # Quick Entry
                    with gr.Tab("‚ö° Quick Entry"):
                        gr.Markdown("""
                        ### Smart Expense Entry
                        Just type naturally! Examples:
                        - `Groceries 2500`
                        - `Movie tickets 800`
                        - `Electricity bill 1500`

                        The AI will automatically categorize it!
                        """)

                        quick_expense_input = gr.Textbox(
                            label="Quick Entry",
                            placeholder="e.g., Lunch 250",
                            lines=1
                        )
                        quick_add_btn = gr.Button("‚ûï Add Expense", variant="primary")
                        quick_expense_output = gr.Markdown()

                    # Detailed Entry
                    with gr.Tab("üìù Detailed Entry"):
                        with gr.Row():
                            expense_desc = gr.Textbox(label="Description", placeholder="e.g., Supermarket groceries")
                            expense_amount = gr.Number(label="Amount (‚Çπ)", value=0, minimum=0)

                        expense_category = gr.Dropdown(
                            label="Category",
                            choices=[
                                "Food & Groceries", "Rent", "Utilities", "Transportation",
                                "Healthcare", "Education", "Entertainment", "Clothing",
                                "Insurance", "EMI/Loans", "Communication", "Personal Care",
                                "Household Items", "Gifts & Donations", "Travel", "Investments",
                                "Emergency Fund", "Other"
                            ],
                            value="Food & Groceries"
                        )

                        detailed_add_btn = gr.Button("‚ûï Add Expense", variant="primary")
                        detailed_expense_output = gr.Markdown()

                # Expense Table
                gr.Markdown("### Recent Expenses")
                expense_table = gr.Dataframe(
                    label="Your Expenses (Last 10)",
                    headers=["Date", "Description", "Category", "Amount (‚Çπ)"],
                    interactive=False
                )

                # Connect buttons
                quick_add_btn.click(
                    fn=add_expense_quick,
                    inputs=[quick_expense_input],
                    outputs=[quick_expense_output, expense_table, quick_expense_input]
                )

                detailed_add_btn.click(
                    fn=add_expense_detailed,
                    inputs=[expense_desc, expense_amount, expense_category],
                    outputs=[detailed_expense_output, expense_table]
                )

            # ============================================================
            # TAB 5: ANALYSIS
            # ============================================================
            with gr.Tab("üìä Analysis", id=4):
                gr.Markdown("## Financial Analysis & Insights")

                analyze_btn = gr.Button("üîç Analyze My Finances", variant="primary", size="lg")

                with gr.Row():
                    with gr.Column():
                        analysis_summary = gr.Markdown(label="Analysis Report")

                with gr.Row():
                    chart1 = gr.Plot(label="Income vs Expenses")
                    chart2 = gr.Plot(label="Expense Breakdown")

                with gr.Row():
                    chart3 = gr.Plot(label="Savings Rate")

                analyze_btn.click(
                    fn=analyze_budget,
                    outputs=[analysis_summary, chart1, chart2, chart3]
                )

                gr.Markdown("""
                ---
                ### üìà Understanding Your Report

                - **Surplus:** You're saving money! Great job!
                - **Deficit:** You're overspending. Time to cut costs.
                - **Savings Rate:** Aim for at least 20% of income
                - **50-30-20 Rule:** 50% needs, 30% wants, 20% savings
                """)

            # ============================================================
            # TAB 6: AI CHAT
            # ============================================================
            with gr.Tab("üí¨ AI Assistant", id=5):
                gr.Markdown("""
                ## Chat with Your AI Financial Advisor

                Ask me anything about your finances! Examples:
                - How can I save more money?
                - Where should I cut expenses?
                - Am I spending too much on entertainment?
                - What investments should I consider?
                """)

                chatbot = gr.Chatbot(height=500)
                msg = gr.Textbox(
                    label="Your Question",
                    placeholder="Type your question here...",
                    lines=2
                )

                with gr.Row():
                    submit_btn = gr.Button("Send", variant="primary")
                    clear_btn = gr.Button("Clear Chat")

                def user(user_message, history):
                    return "", history + [[user_message, None]]

                def bot(history):
                    user_message = history[-1][0]
                    bot_message = chatbot_response(user_message, history)
                    history[-1][1] = bot_message
                    return history

                msg.submit(user, [msg, chatbot], [msg, chatbot], queue=False).then(
                    bot, chatbot, chatbot
                )
                submit_btn.click(user, [msg, chatbot], [msg, chatbot], queue=False).then(
                    bot, chatbot, chatbot
                )
                clear_btn.click(lambda: None, None, chatbot, queue=False)

                gr.Markdown("""
                ---
                ### üí° Conversation Tips
                - Be specific about your questions
                - Mention time frames (this month, this year)
                - Ask for comparisons or alternatives
                - Request step-by-step advice
                """)

            # ============================================================
            # TAB 7: SUGGESTIONS
            # ============================================================
            with gr.Tab("üí° Suggestions", id=6):
                gr.Markdown("## Expert Recommendations")

                with gr.Tabs():
                    # Tax Savings
                    with gr.Tab("üí∞ Tax Savings"):
                        tax_btn = gr.Button("Get Tax Saving Suggestions", variant="primary")
                        tax_output = gr.Markdown()

                        tax_btn.click(
                            fn=get_tax_suggestions,
                            outputs=[tax_output]
                        )

                    # Government Schemes
                    with gr.Tab("üèõÔ∏è Government Schemes"):
                        schemes_btn = gr.Button("Show Applicable Schemes", variant="primary")
                        schemes_output = gr.Markdown()

                        schemes_btn.click(
                            fn=get_government_schemes,
                            outputs=[schemes_output]
                        )

                    # Peer Comparison
                    with gr.Tab("üìä Peer Comparison"):
                        compare_btn = gr.Button("Compare with Peers", variant="primary")
                        compare_output = gr.Markdown()

                        compare_btn.click(
                            fn=compare_with_peers,
                            outputs=[compare_output]
                        )

            # ============================================================
            # TAB 8: DAILY TRACKING
            # ============================================================
            with gr.Tab("üìÖ Daily Tracker", id=7):
                gr.Markdown("""
                ## Quick Daily Check-in
                ### Takes only 30 seconds! ‚ö°
                """)

                with gr.Group():
                    gr.Markdown("### Did you spend money today?")

                    with gr.Row():
                        daily_spent = gr.Radio(
                            choices=["Yes", "No"],
                            label="",
                            value="Yes"
                        )

                    with gr.Column(visible=True) as daily_details:
                        gr.Markdown("### Quick estimate:")
                        daily_amount = gr.Radio(
                            choices=[
                                "Less than ‚Çπ200 (minimal)",
                                "‚Çπ200-500 (moderate)",
                                "‚Çπ500-1000 (average)",
                                "More than ‚Çπ1000 (high)"
                            ],
                            label="How much did you spend?",
                            value="‚Çπ200-500 (moderate)"
                        )

                        daily_category = gr.Dropdown(
                            label="Main spending category",
                            choices=["Food", "Transport", "Shopping", "Bills", "Entertainment", "Other"],
                            value="Food"
                        )

                    daily_submit = gr.Button("‚úÖ Log Today's Expenses", variant="primary", size="lg")
                    daily_output = gr.Markdown()

                def log_daily_expense(spent, amount_range, category):
                    if spent == "No":
                        app_state.gamification_points += 50
                        return f"""
üéâ **No-spend day!**

Congratulations! You earned 50 points! üåü

üíé Total Points: {app_state.gamification_points}

Keep up the great discipline!
"""

                    # Parse amount
                    amount_map = {
                        "Less than ‚Çπ200 (minimal)": 100,
                        "‚Çπ200-500 (moderate)": 350,
                        "‚Çπ500-1000 (average)": 750,
                        "More than ‚Çπ1000 (high)": 1500
                    }
                    amount = amount_map.get(amount_range, 500)

                    # Add to expenses
                    expense_entry = {
                        'description': f'Daily {category}',
                        'amount': amount,
                        'category': category,
                        'date': datetime.now().strftime("%Y-%m-%d"),
                        'time': datetime.now().strftime("%H:%M:%S")
                    }
                    app_state.expenses_list.append(expense_entry)

                    return f"""
‚úÖ **Logged successfully!**

- Amount: ‚Çπ{amount}
- Category: {category}
- Date: {datetime.now().strftime("%Y-%m-%d")}

Great job tracking! Consistency is key to financial awareness! üí™
"""

                daily_submit.click(
                    fn=log_daily_expense,
                    inputs=[daily_spent, daily_amount, daily_category],
                    outputs=[daily_output]
                )

                gr.Markdown("""
                ---
                ### üéØ Daily Tracking Benefits
                - Build awareness of spending habits
                - Takes only 30 seconds per day
                - Automatic categorization
                - Earn points for no-spend days!
                """)

            # ============================================================
            # TAB 9: GAMIFICATION
            # ============================================================
            with gr.Tab("üéÆ Challenges", id=8):
                gr.Markdown("## Financial Challenges & Rewards")

                def show_current_points():
                    return f"""
# üíé Your Progress

## Current Points: {app_state.gamification_points}

### Available Challenges:

#### ü•â Easy Challenges (300 points)
- **Save ‚Çπ1000 This Week**
  - Track all expenses for 7 days
  - Reduce spending by ‚Çπ1000

#### ü•à Medium Challenges (500 points)
- **No Dining Out Week**
  - Cook at home for 7 days straight
  - Share your progress daily

- **5-Day Packed Lunch Challenge**
  - Bring lunch from home
  - Save on food expenses

#### ü•á Hard Challenges (1000 points)
- **Reduce Utility Bills by 20%**
  - Track usage for 30 days
  - Implement saving strategies

- **Monthly Savings Goal**
  - Save 30% of income this month
  - No unnecessary purchases

---

### üèÜ Achievements
- üåü **First Week**: Complete profile + 7 days tracking
- üí∞ **Saver**: Save money for 30 days straight
- üìä **Analyst**: Review budget weekly for a month
- üéØ **Goal Getter**: Achieve one financial goal

---

### Rewards Tier
- 0-500 points: ü•â Bronze
- 501-1500 points: ü•à Silver
- 1501-3000 points: ü•á Gold
- 3000+ points: üíé Platinum

**Your current tier:** {'ü•â Bronze' if app_state.gamification_points < 500 else 'ü•à Silver' if app_state.gamification_points < 1500 else 'ü•á Gold' if app_state.gamification_points < 3000 else 'üíé Platinum'}
"""

                points_display = gr.Markdown()
                refresh_points = gr.Button("üîÑ Refresh Points")

                refresh_points.click(
                    fn=show_current_points,
                    outputs=[points_display]
                )

                # Initialize display
                app.load(
                    fn=show_current_points,
                    outputs=[points_display]
                )

                gr.Markdown("""
                ---
                ### üí° How to Earn Points
                - Complete profile: 100 points
                - Add income sources: 50 points each
                - Track expenses daily: 10 points/day
                - No-spend day: 50 points
                - Complete challenges: 300-1000 points
                - Achieve savings goal: 500 points
                """)

            # ============================================================
            # TAB 10: REPORTS
            # ============================================================
            with gr.Tab("üìÑ Reports", id=9):
                gr.Markdown("## Export Your Financial Reports")

                with gr.Group():
                    gr.Markdown("### Generate Comprehensive Report")

                    report_type = gr.Radio(
                        choices=["Monthly Summary", "Detailed Analysis", "Tax Report", "Full Report"],
                        label="Report Type",
                        value="Monthly Summary"
                    )

                    generate_report_btn = gr.Button("üì• Generate Report", variant="primary")
                    report_output = gr.Markdown()

                def generate_report(report_type):
                    if not app_state.analysis:
                        return "‚ö†Ô∏è Please complete budget analysis first"

                    report = f"""
# üìä Financial Report
## {report_type}
### Generated on: {datetime.now().strftime("%Y-%m-%d %H:%M:%S")}

---

## üë§ Profile Summary
- **Name:** {app_state.user_profile.get('name', 'N/A')}
- **Age:** {app_state.user_profile.get('age', 'N/A')}
- **Category:** {app_state.user_profile.get('category', 'N/A')}
- **Location:** {app_state.user_profile.get('city', 'N/A')}, {app_state.user_profile.get('country', 'N/A')}

---

## üí∞ Financial Summary

### Income
- **Total Monthly Income:** ‚Çπ{app_state.analysis['total_income']:,.0f}
- **Number of Sources:** {len(app_state.income_list)}

### Expenses
- **Total Monthly Expenses:** ‚Çπ{app_state.analysis['total_expenses']:,.0f}
- **Number of Transactions:** {len(app_state.expenses_list)}

### Savings
- **Monthly Savings:** ‚Çπ{app_state.analysis['savings']:,.0f}
- **Savings Rate:** {app_state.analysis['savings_rate']:.1f}%
- **Status:** {'‚úÖ Surplus' if app_state.analysis['status'] == 'surplus' else '‚ö†Ô∏è Deficit'}

---

## üìä Expense Breakdown
"""

                    for category, amount in app_state.analysis.get('category_breakdown', {}).items():
                        percentage = (amount / app_state.analysis['total_expenses'] * 100) if app_state.analysis['total_expenses'] > 0 else 0
                        report += f"\n- **{category}:** ‚Çπ{amount:,.0f} ({percentage:.1f}%)"

                    report += f"""

---

## üéØ Goals
"""
                    for i, goal in enumerate(app_state.user_profile.get('goals', []), 1):
                        report += f"\n{i}. {goal}"

                    report += f"""

---

## üí° Recommendations

1. **Savings Goal:** Aim to save at least 20% of income
2. **Emergency Fund:** Build 6 months of expenses (‚Çπ{app_state.analysis['total_expenses'] * 6:,.0f})
3. **Investment:** Consider SIP in mutual funds with saved amount
4. **Tax Planning:** Utilize Section 80C for tax savings

---

## üìà Progress Tracking
- **Gamification Points:** {app_state.gamification_points}
- **Days Tracked:** {len(set(e['date'] for e in app_state.expenses_list))}
- **Consistency Score:** {'Excellent' if len(app_state.expenses_list) > 20 else 'Good' if len(app_state.expenses_list) > 10 else 'Getting Started'}

---

*This report is generated by AI Financial Advisor*
*For best results, update your data regularly*
"""

                    return report

                generate_report_btn.click(
                    fn=generate_report,
                    inputs=[report_type],
                    outputs=[report_output]
                )

                gr.Markdown("""
                ---
                ### üìã Report Features
                - Comprehensive financial overview
                - Category-wise expense analysis
                - Savings rate tracking
                - Goal progress monitoring
                - Actionable recommendations

                üí° **Tip:** Generate reports weekly to track your progress!
                """)

        # Footer
        gr.Markdown("""
        ---
        <div style='text-align: center; padding: 20px;'>
            <p><strong>AI Financial Advisor</strong> | Powered by IBM Granite 3.2-2B</p>
            <p>üîí Your data is private and secure | üí° All AI suggestions are for informational purposes</p>
        </div>
        """)

    return app

# ============================================================================
# LAUNCH APPLICATION
# ============================================================================

if __name__ == "__main__":
    print("""
    ‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó
    ‚ïë                                                              ‚ïë
    ‚ïë          ü§ñ AI FINANCIAL ADVISOR - GRADIO UI ü§ñ             ‚ïë
    ‚ïë                                                              ‚ïë
    ‚ïë     Powered by IBM Granite 3.2-2B Instruct Model            ‚ïë
    ‚ïë                                                              ‚ïë
    ‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù

    üöÄ Starting Gradio Interface...
    üì± The app will open in a new browser tab
    üåê You'll get a public URL to share (optional)

    """)

    # Create and launch the interface
    demo = create_ui()

    # Launch with public URL for Colab
    demo.launch(
        share=True,  # Creates public URL for sharing
        debug=True,  # Enable debug mode
        show_error=True  # Show detailed errors
    )

    print("""
    ‚úÖ Application is running!
    üì± Access it through the link above
    üõë To stop: Click the stop button or press Ctrl+C
    """)
```
